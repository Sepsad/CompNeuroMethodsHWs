```python
import numpy as np
import matplotlib.pyplot as plt

m_diff = np.linspace(-5, 5, 200)  # range of (m2 - m1)

plt.figure(figsize=(7,5))

# Different combinations of line styles and markers
styles = [
    {'linestyle': '-', 'marker': 'o', 'markevery': 30, 'markersize': 6},
    {'linestyle': '--', 'marker': 's', 'markevery': 30, 'markersize': 6},
    {'linestyle': ':', 'marker': '^', 'markevery': 30, 'markersize': 6},
    {'linestyle': '-.', 'marker': 'x', 'markevery': 30, 'markersize': 6},
    {'linestyle': '-', 'marker': 'd', 'markevery': 30, 'markersize': 6}
]

for i, beta in enumerate([-10,-1,0.1, 1, 10]):
    p = 1 / (1 + np.exp(beta * (m_diff)))
    plt.plot(m_diff, p, label=f'beta = {beta}', color='black', **styles[i])

plt.axhline(0.5, color='k', linestyle='--', alpha=0.5)
plt.xlabel('m2 - m1 = d')
plt.ylabel('p(c=1)')
plt.title('Logistic Curve for Different β')
plt.legend()
plt.grid(True)
plt.savefig('./logistic_curve.png')
plt.show()


```
